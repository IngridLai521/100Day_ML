### 1. 模型的泛化能力 (generalization) 是指什麼？ 
### 2. 分類問題與回歸問題分別可用的目標函數有哪些？

1. 泛化能力
指的是一個機器學習模型遇到新樣本時的表現能力，泛化能力越高代表對樣本外的資料適應能力越好。

2. 分類與迴歸問題可用的目標函數
分類問題常用的損失函數為交叉熵(cross-entropy)，迴歸問題常使用的則有MSE、MAE，另外也可以參考這個網站的內容：https://www.jiqizhixin.com/articles/2018-06-21-3
